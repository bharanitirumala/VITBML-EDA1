            U.S. Air Pollution Data summary
This dataset deals with pollution in the U.S. Pollution in the U.S. has been well documented by the U.S. EPA.

Includes four major pollutants (Nitrogen Dioxide, Sulphur Dioxide, Carbon Monoxide and Ozone).

State Code : The code allocated by US EPA to each state
County code : The code of counties in a specific state allocated by US EPA
Site Num : The site number in a specific county allocated by US EPA
Address: Address of the monitoring site
State : State of monitoring site
County : County of monitoring site
City : City of the monitoring site
Date Local : Date of monitoring
The four pollutants (NO2, O3, SO2 and O3) each has 5 specific columns. For instance, for NO2:

NO2 Units : The units measured for NO2
NO2 Mean : The arithmetic mean of concentration of NO2 within a given day
NO2 AQI : The calculated air quality index of NO2 within a given day
NO2 1st Max Value : The maximum value obtained for NO2 concentration in a given day
NO2 1st Max Hour : The hour when the maximum NO2 concentration was recorded in a given day 
                

```
import numpy as np
```

```
import pandas as pd
```

```
uspollution=pd.read_csv('/content/drive/My Drive/us air pollution data.csv')
print(uspollution)
```

```
uspollution.dropna(axis=0,inplace=True)
print(uspollution)
```

```
uspollution.columns
```

```
import seaborn as sns
```

```
sns.boxplot(x="City",y='NO2 1st Max Value',data=uspollution);
```

```
from matplotlib import pyplot as pt
```

```
uspollution.tail()
```

```
v=uspollution.groupby('NO2 1st Max Value')['NO2 1st Max Hour'].mean()
pt.style.use('ggplot')
v.plot(kind='hist',color='blue')
pt.show()
```

```
pt.scatter(x=uspollution['SO2 1st Max Value'],y=uspollution['NO2 1st Max Value'],c='red');
```

```
r=uspollution.groupby('City')['NO2 1st Max Hour','NO2 1st Max Value'].mean();
r.plot(kind='bar')
pt.title('Comparison of City & NO2')
pt.xlabel('City')
pt.ylabel('NO2')
```

```
r=uspollution.groupby('City')['NO2 1st Max Hour'].mean();
r.plot(kind='pie')
pt.title('Comparison of City & NO2')
```

```
sns.violinplot(x="City",y='NO2 1st Max Value',data=uspollution);
```

```
sns.swarmplot(x="City",y='NO2 1st Max Value',data=uspollution);
```

```
sns.stripplot(x="City",y='NO2 1st Max Value',data=uspollution);
```

```
n=uspollution.groupby('City')['NO2 1st Max Value'].mean()
n.plot()
pt.xlabel('City')
pt.ylabel('NO2 1st Max Hour')
pt.title('Air pollution')
pt.show
```
